```python
# The size column contains values in different formats, such as '1.0M', '2.5k', etc.
# We need to convert these values to bytes for consistency
```

```python
def size_process(item):
    if str(item)[-1] =="M":
        res = float(str(item).replace("M",""))
        return res
    
    elif str(item)[-1] =="K":
        res = float(str(item).replace("K",""))
        res = res / 1024
        return round(res,3)
    else:
        return str(np.nan)
```

```python
df["size"]= df["size"].apply(size_process) # apply the function to Size column
```

## Product Strategy & Portfolio Insights

Objective: Help product managers understand the current app landscape.

### 1. Top Categories by Number of Apps

Goal: Identify the most saturated and competitive categories.

```python
# get the top 10 categories by number of apps

category_counts =df['category'].value_counts().head(10) 

# Create a horizontal bar plot for the top 10 categories
plt.figure(figsize=(12, 8))
category_counts.plot(kind='barh', color='skyblue')

plt.title('Top 10 Categories by Number of Apps', fontsize=16, fontweight='bold')
plt.xlabel('Number of Apps', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.ylabel('Category',fontsize=14, fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the most popular category on top
plt.tight_layout()
plt.show()
```

### 2. Top Genres by Average Rating.

Goal: Find the top 10 genres with the highest average user rating.



```python
# Group by 'genres'and calculate the average rating
top_genres = df.groupby('genres')['rating'].mean().sort_values(ascending=False).head(10)

# Create a horizontal bar plot for the top 10 genres by average rating
plt.figure(figsize=(12, 8))
top_genres.plot(kind='barh', color='mediumseagreen')
plt.title('Top 10 Genres by Average Rating', fontsize=16, fontweight='bold')
plt.xlabel('Average Rating', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.ylabel('Genre', fontsize=14, fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest rated genre on top
plt.tight_layout()
plt.show()
```

### 3. Top content ratings (e.g., Everyone, Teen) by install count

Goal:Group apps by their content_rating (e.g., 'Everyone', 'Teen', 'Mature 17+'), and sum up the total number of installs for each.

```python
# Group by content rating and sum installs

install_by_content_rating = df.groupby('content_rating')['installs'].sum().sort_values(ascending=False)

# Create a horizontal bar plot for installs by content rating
plt.figure(figsize=(12, 8))
install_by_content_rating.plot(kind='barh', color='salmon')
plt.title('Total Installs by Content Rating', fontsize=16, fontweight='bold')
plt.xlabel('Total Installs', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.ylabel('Content Rating', fontsize=14, fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest installs
plt.tight_layout()
plt.show()
```

#### Insight Summary

+ “Everyone” dominates with over 52 billion installs, confirming that the vast majority of Android apps are targeted at all age groups.

+ “Teen” rated apps also have strong traction, with over 16 billion installs, indicating a sizable market for social, gaming, and creative tools aimed at younger users.

+ Install counts sharply decline after that — with “Everyone 10+” and “Mature 17+” trailing far behind.

+ “Adults only 18+” and “Unrated” apps have almost no installs in comparison, likely due to store restrictions, niche audiences, or limited visibility.

### 3. Top Content Ratings by Install Count
- **Everyone-rated** apps dominate with over **52 billion installs**.
- **Teen** apps follow with ~16 billion installs.
- Apps rated **Adults only 18+** or **Unrated** have extremely low install counts.

---

### 4. Categories with Most Paid Apps
- **FAMILY**, **GAME**, **TOOLS**, and **MEDICAL** have the most paid apps.
- These categories combine **trust**, **entertainment**, and **utility**, making users more willing to pay.

---

### 5. Genres That Perform Best (Rating vs. Installs)
- High **ratings**: **Events**, **Education**, **Art & Design**
- High **installs**: **Communication**, **Video Players**, **Social**
- Key insight: **Popular ≠ Loved**. There’s a quality gap in mass-market genres.

---

## Common Strategic Themes Across Analyses

### 1. Popularity vs. Quality Trade-Off
- High installs often come with **lower ratings**.
- High ratings appear in **niche, purpose-driven genres**.

> **Strategy:** Invest in UX for mass-market apps, and visibility for niche, high-quality apps.

---

### 2. Trusted Categories Monetize Better
- **FAMILY**, **TOOLS**, and **MEDICAL** support more paid apps.
- Users are willing to pay in **high-trust**, **ad-sensitive**, or **professional-grade** contexts.

---

### 3. Target “Everyone” and “Teen” Audiences
- Broadest reach, best visibility in the Play Store.
- **Missing or incorrect content ratings result in near-zero installs.**

---

### 4. Emerging High-Impact Niches
- **Parenting**, **Events**, **Education**, and **Productivity** offer strong user loyalty.
- Ideal for apps focusing on **long-term retention**, **subscription models**, or **freemium upsells**.

---

## Summary for Product Teams

If you're launching or optimizing an app:
- Choose **“Everyone”** or **“Teen”** as your audience unless highly niche.
- Optimize for **either reach (install)** or **engagement (ratings)** — rarely both.
- Consider paid models in **trusted categories** where users expect quality.
- Invest in visibility for **underrated, high-rated genres** — they’re ripe for disruption.

### 3: Average Price by Category (Paid Apps Only)

Goal:Identify categories where developers successfully charge higher average prices.

```python
# Filter Paid Apps Only
paid_apps = df[df['type'] == 'Paid']

# Group by Category and Compute Average Price
avg_price_by_category = paid_apps.groupby('category')['price_usd'].mean().sort_values(ascending=False)

# Create a horizontal bar plot for average price by category
plt.figure(figsize=(12, 8))
avg_price_by_category.plot(kind='barh', color='mediumpurple')
plt.title('Average Price of Paid Apps by Category', fontsize=16, fontweight='bold')
plt.xlabel('Average Price in USD', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.ylabel('Category', fontsize=14, fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest average price on top
plt.tight_layout()
plt.show()
```

#### Insight:
+ Categories like Finance, Lifestyle, and Events show significantly higher average prices, with Finance leading by a wide margin (over $160 on average — likely due to pricing outliers).

+ Other premium categories include Business, Medical, and Productivity, where users often expect professional-grade features.

+ Mass-market categories such as Games, Entertainment, and Tools have low average prices, typically under $2 — optimized for user volume and accessibility.

#### Strategic Implications:
+ Monetize more aggressively in high-trust, high-value categories like Finance, Business, and Medical using premium or subscription models.

+ Use low-cost or freemium pricing for consumer-oriented categories like Games and Entertainment to maximize reach and engagement.

+ Watch out for outliers that skew average price calculations — consider using median pricing when making pricing decisions.

### 4: Revenue Potential Estimate

Goal:Identify which apps (or categories) generate the highest revenue potential, assuming all installs were paid.

```python
paid_apps = df[df['type'] == 'Paid']

# Calculate Revenue Potential Estimate
paid_apps['revenue_potential'] = paid_apps['price_usd'] * paid_apps['installs']

# Top 10 Revenue-Generating Apps 

top_revenue_apps = paid_apps[['app', 'category', 'price_usd', 'installs', 'revenue_potential']]
top_revenue_apps = top_revenue_apps.sort_values(by='revenue_potential', ascending=False).head(10)

# Revenue Potential Estimate by Category

revenue_by_category= paid_apps.groupby('category')['revenue_potential'].sum().sort_values(ascending=False)

# # Create a horizontal bar plot 

plt.figure(figsize=(12, 8))
revenue_by_category.head(10).plot(kind='barh', color='darkcyan')
plt.title('Top 10 Categories by Estimated Revenue', fontsize=16, fontweight='bold')
plt.xlabel('Estimated Revenue (USD)', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.ylabel('Category', fontsize=14, fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest average price on top
plt.tight_layout()
plt.show()



```

#### Key Insight:
+ The FAMILY category generates the highest estimated revenue among paid apps — likely due to high install volume, even with modest pricing.

+ LIFESTYLE and GAME categories also show strong revenue performance, driven by broad appeal and solid user bases.

+ FINANCE, despite having the highest average prices, ranks lower in total estimated revenue — suggesting fewer installs or more niche usage.

+ Utility-focused categories like PHOTOGRAPHY, TOOLS, and MEDICAL offer moderate but consistent revenue streams.

#### Strategic Implications:
+ Maximize revenue by targeting categories with a strong user base + acceptable pricing — not just high prices alone.

+ For categories like FINANCE, ensure pricing is supported by a clear premium value proposition to encourage adoption.

+ Consider FAMILY, GAMES, and LIFESTYLE as ideal for paid apps aiming for both scale and earnings.

+ Utility categories remain viable for sustainable monetization, especially with ad-free or feature-rich versions.

**Recommendation:**

When planning a monetization strategy, prioritize app categories where you can combine:

+ Consistent user demand

+ Mid-tier pricing ($1.99–$9.99)

+ Strong functional or emotional value

### 5: Correlation Between Price, Rating, and Installs

Quantify the relationship between:

 + price_usd and rating

 + price_usd and installs

### 1. Categories with High Average Price but Low Average Installs

```python
# Filter paid apps

paid_apps = df[df['price_usd'] > 0]

#Group by Category
category_price_installs = paid_apps.groupby('category').agg({
    'price_usd': 'mean',
    'installs': 'mean',
    'app': 'count'
}).rename(columns={
    'price_usd': 'avg_price',
    'installs': 'avg_installs',
    'app': 'num_paid_apps'
}).sort_values(by='avg_price', ascending=False)

#Filter for Low Installs with High Price

median_installs = category_price_installs['avg_installs'].median()

# Filter categories with high price but low installs
overpriced_categories = category_price_installs[
    (category_price_installs['avg_price'] > 10) & 
    (category_price_installs['avg_installs'] < median_installs)]

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
overpriced_categories.sort_values(by='avg_price', ascending=False)['avg_price'].plot(
    kind='barh', color='firebrick', edgecolor='black')
plt.title('Categories with High Price but Low Average Installs', fontsize=16, fontweight='bold')
plt.xlabel('Average Price (USD)',fontsize=14, fontweight='bold')
plt.ylabel('Category', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

### Insight: 
+ The following categories have significantly high average app prices but low average installs:

  + Finance: ~$170

  + Events: ~$110

  + Business: ~$15

  + Medical: ~$12

+ These apps may be overpriced relative to user demand, signaling pricing inefficiency or niche targeting.

#### Strategic Implications:
+ Monetization Risk:High pricing in low-demand categories could be limiting adoption. Users may be deterred by high upfront costs without clear perceived value.

+ Targeting or UX Gaps:In categories like Finance and Medical, where trust and usability are critical, low installs may also indicate poor UX or lack of trust.

+ Optimization Opportunity:Re-evaluating pricing strategies in these categories could help:

  + Boost adoption (via freemium or tiered pricing)

  + Improve ROI by aligning with customer expectations

  + Identify apps that may be better suited to a subscription model

Recommendation:
App publishers in these categories should run pricing experiments (e.g., A/B testing, discounts) and explore value communication improvements (clearer benefits, free trials, etc.).

### 2. Identify Underrated Apps

Goal: Find apps that are loved by users (high rating), but haven’t yet reached a large audience (low installs).

```python
# Define thresholds
high_rating_threshold = 4.5
low_installs_threshold = df['installs'].quantile(0.25)  # bottom 25% in installs

# Filter apps
underrated_apps = df[
    (df['rating'] >= high_rating_threshold) &
    (df['installs'] <= low_installs_threshold)
]

# Select relevant columns and sort by rating
underrated_top = underrated_apps.sort_values(by='rating', ascending=False)
underrated_top[['app', 'category', 'rating', 'installs']].head(10)

# Bar plot for Underrated Apps

top_underrated = underrated_top.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_underrated, x='rating', y='app', palette='viridis')
plt.title('Top 10 Underrated Apps (High Rating, Low Installs)', fontsize=16, fontweight='bold')
plt.xlabel('Rating', fontsize=14, fontweight='bold')
plt.ylabel('App Name', fontsize=14, fontweight='bold')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### Insight:
+ All top 10 apps have ratings near 5.0, reflecting exceptional user satisfaction.

+ Despite excellent ratings, these apps have very low install counts, indicating:

    + Lack of discovery or marketing

    + Possibly limited targeting or visibility in app store searches

+ Apps cover diverse domains: education (R Programming Tutorial, Chemistry), religion (Quran app), task tracking, and niche quizzes.

### 3. Identify Very Old Apps with High Installs

Goal: Find apps with high install counts despite not being updated for a long time.

```python
# Filter apps not updated in 2+ years
legacy_apps = df[df['days_since_update'] > 730]

# Sort by highest installs
top_legacy_installs = legacy_apps.sort_values(by='installs', ascending=False)

# Show top 10
top_legacy_installs[['app', 'category', 'last_updated', 'days_since_update', 'installs', 'rating']].head(10)

# Top 10 apps
top10_legacy = top_legacy_installs.head(10)

# Bar plot for Top 10 Legacy Apps by Installs
plt.figure(figsize=(12, 8))
sns.barplot(data=top10_legacy, x='installs', y='app', palette='plasma')
plt.title('Top 10 Legacy Apps (Not Updated in 2+ Years)', fontsize=16, fontweight='bold')
plt.xlabel('Installs', fontsize=14, fontweight='bold')
plt.ylabel('App Name', fontsize=14, fontweight='bold')
plt.xticks(rotation=90,fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Insight:
+ Apps like Facebook, Gmail, Google+, Google Photos, and Hangouts top the list of legacy apps with over 1 billion installs — despite not being updated in more than 2 years.

+ These are mostly pre-installed or default apps by major vendors (Meta, Google), showing massive legacy footprints.

+ Despite being outdated, they maintain a massive user base due to historical installs and device bundling.
